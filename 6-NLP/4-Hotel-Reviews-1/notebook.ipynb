```python
import pandas as pd
import time

df = pd.read_csv("../data/Hotel_Reviews.csv")
df.describe()
```

```python
df["Reviewer_Nationality"].value_counts()
```

```python
df["Hotel_Name"].value_counts()
```

```python
df["Calc_Score"] = round(df.groupby(["Hotel_Name"]).Reviewer_Score.transform("mean"), 1)
```

```python
print(f"No negative review: {sum(df['Negative_Review'] == 'No Negative')}")
print(f"No positive review: {sum(df['Positive_Review'] == 'No Positive')}")
print(f"No review: {sum((df['Negative_Review'] == 'No Negative') & (df['Positive_Review'] == 'No Positive'))}")
```

```python
df.drop(["lat", "lng"], axis="columns")
```

```python
def replace_address(row):
    if "Netherlands" in row["Hotel_Address"]:
        return "Amsterdam, Netherlands"
    elif "Barcelona" in row["Hotel_Address"]:
        return "Barcelona, Spain"
    elif "United Kingdom" in row["Hotel_Address"]:
        return "London, United Kingdom"
    elif "Milan" in row["Hotel_Address"]:        
        return "Milan, Italy"
    elif "France" in row["Hotel_Address"]:
        return "Paris, France"
    elif "Vienna" in row["Hotel_Address"]:
        return "Vienna, Austria"
    
df["Hotel_Address"] = df.apply(replace_address, axis = "columns")

df["Hotel_Address"].value_counts()
```

```python
df.drop(["Additional_Number_of_Scoring", "Review_Total_Negative_Word_Counts", "Review_Total_Positive_Word_Counts", "Review_Date", "days_since_review", "Total_Number_of_Reviews_Reviewer_Has_Given"], axis = 1, inplace=True)
df.Total_Number_of_Reviews = df.groupby('Hotel_Name')["Hotel_Name"].transform('count')
df.Average_Score = round(df.groupby("Hotel_Name").Reviewer_Score.transform("mean"), 1)
```

```python
df.Tags.str.strip("[']")
df.Tags = df.Tags.str.replace(" ', '", ",", regex = False)
```

```python
tags = ["Leisure trip", "Couple", "Solo_traveler", "Business_trip", "Family_with_young_children", "Family_with_older_children", "With_a_pet"]

for t in tags:
    df[t.replace(" ", "_")] = df.Tags.apply(lambda tag: 1 if t in tag else 0)

df["Group"] = df.Tags.apply(lambda tag: 1 if "Group" in tag or "Travelers with friends" in tag else 0)
```

```python
df.to_csv(r'../data/Hotel_Reviews_Filtered.csv', index = False)
```
